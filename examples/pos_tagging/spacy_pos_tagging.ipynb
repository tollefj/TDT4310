```python
import spacy
import nltk
nlp = spacy.load("en_core_web_sm")
```

```python
corpus = nltk.corpus.brown.sents()[:5]
```

```python
import re

corpus = nltk.corpus.brown.sents()[:5]
def preprocess(sent):
    sent = " ".join(sent)
    sent = re.sub(r"[^\w,.!?]", " ", sent)
    sent = re.sub(r"\s+", " ", sent)
    return sent.strip()
corpus = [preprocess(sent) for sent in corpus]
corpus
```

```python
words, tags = [], []
start_words = []

for doc in nlp.pipe(corpus, batch_size=1000):
    _words, _tags = zip(*[(t.text, t.pos_) for t in doc])
    start_words.append(_words[0])
    # _words = [w.lower() for w in _words if w.isalnum()]
    words.extend(_words)
    tags.extend(_tags)

for w, t in zip(words, tags):
    print(w, t)
```
